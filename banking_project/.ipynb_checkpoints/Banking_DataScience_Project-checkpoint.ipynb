# Classification Task

The Classification class contains five methods. The first method removes the columns and rows the NaN entries. The removal of such entries for the correct prediction. The "preprocessing" method encodes the labels. The "training" method does the classification task and return two dict objects that contains ROC "Receiver Operating Curve" and confusion matrix data for the applied algorithms. The last two methods plots ROC curve and confusion matrix. 

```python
class Classification:
    def __init__(self, df):
        df= df.dropna(how='all', axis=1)
        df = df.dropna(how='all')
    
    def preprocessing(self, df):
        le_s=preprocessing.LabelEncoder()
        for i in df.columns:
            if(df[i].dtype==object):
                df[i]=le_s.fit_transform(df[i])
        return df
    #########################################################################
    def training(self, data,labels, X_test, y_test, algo_list):
    
        cv = KFold(len(labels), n_folds=10)
        classifiers={}
        confusion={}
        for i in algo_list:
            for traincv, testcv in cv:
                proba= algo_list.get(i).fit(data.iloc[traincv], labels.iloc[traincv])
                #probas_=proba.predict(data.iloc[testcv])
                #print f1_score(labels.iloc[testcv], probas_)  
            
            probas_=proba.predict_proba(X_test)
            confusion[i]=confusion_matrix(y_test, proba.predict(X_test))
            classifiers[i]=roc_curve(y_test, probas_[:, 1])
            
        return classifiers, confusion
    
    ######################
    def plot_roc_curve(self, roc_auc):
        
        pl.figure(figsize=(15,10))
        pl.plot([0, 1], [0, 1], 'k--')
        for i in roc_auc:
            fpr,tpr,_= roc_auc.get(i)
            roc=auc(fpr,tpr)
            pl.plot(fpr, tpr, label='ROC curve %s (testing data) (area = %0.4f)' % (i, roc))
        pl.xlabel('False Positive Rate', fontsize=20)
        pl.ylabel('True Positive Rate', fontsize=20)
        pl.title('ROC Curve', fontsize=20)
        pl.legend(loc="lower right", fontsize=20)
    
    def draw_confusion_matrix(self, confusion):
        pl.figure(figsize=(5,15))
        for j,i in enumerate(confusion):
            pl.subplot(4,1,j+1)
            sns.heatmap(confusion.get(i), annot=True, fmt="d")
            pl.tight_layout()
            pl.title(i)
            pl.xlabel("Predicted Value")
            pl.ylabel("Real Value")
        
```

```python
algo_classification=Classification(df)
algo_classification.preprocessing(df)
labels=df["Subscription"]
df=df.drop("Subscription", axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, labels, test_size=0.25, random_state=42)
algo_list={}
algo_list['Decision Tree']=DecisionTreeClassifier()
algo_list['Gaussian Naive']=GaussianNB()
algo_list['Random Forest']=RandomForestClassifier(n_estimators=20)

roc, confusion=algo_classification.training(X_train, y_train,X_test,y_test, algo_list)
algo_classification.plot_roc_curve(roc)
```

```python
algo_classification.draw_confusion_matrix(confusion)
```

# Relationship Between Features

To find relationship between features, following three coefficients are used:

1) Correlation between binary variables (phi coefficient) 2) Correlation between binary and continous variables (point-biserial coefficient) 3) Correlation between continous variables (Pearson correlation coefficient)


```python
corr_matrix=np.zeros((df.shape[1],df.shape[1]))
col = df.columns
for i in range(corr_matrix.shape[0]):
    first=df[col[i]]
    for j in range(corr_matrix.shape[0]):
        second=df[col[j]]
        if len(first.unique())<=2 and len(second.unique())<=2:
            value=matthews_corrcoef(first, second)
        elif (len(first.unique())>2 and len(second.unique())>2):
            value=stats.pearsonr(first, second)[0]
        else:
            value=stats.pointbiserialr(first, second)[0]
        corr_matrix[i,j]=np.nan_to_num(value)

sns.heatmap(pd.DataFrame(corr_matrix, columns=col, index=col), square=True)
```
